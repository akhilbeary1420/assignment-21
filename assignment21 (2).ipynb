```python
bos = pd.DataFrame(boston.data)
bos.head(2)
```

```python
bos.columns = boston.feature_names
bos.head(2)
```

```python
bos['PRICE'] = boston.target
bos.head(2)
```

```python
bos.rename(columns={'CRIM': 'Crime_Rate', 'ZN': 'Lrg_Res_Zones', 'INDUS':'Industry'}, inplace=True)
bos.head(2)
```

```python
sns.set(style='ticks')
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(11.7, 8.27)
sns.regplot(y="PRICE", x="Crime_Rate", data=bos, fit_reg=True)
plt.ylim(0,50) # set Y axis range to minimum of zero
plt.title("Home Price Versus Crime Rate")
```

```python
# create X and y
feature_col = ['Crime_Rate']
#else = bos.drop('PRICE',1) # since we added price, drop it here for everything else
X = bos[feature_col]
y = bos.PRICE

# follow the usual sklearn pattern: import, instantiate, fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X, y)

# print intercept and coefficients
print(lm.intercept_)
print(lm.coef_)
```

```python
# you have to create a DataFrame since the Statsmodels formula interface expects it
X_new = pd.DataFrame({'Crime_Rate': [87]})
X_new.head()
```

```python
# use the model to make predictions on a new value
lm.predict(X_new)
```

```python
X_new = pd.DataFrame({'Crime_Rate': [bos.Crime_Rate.min(), bos.Crime_Rate.max()]})
X_new.head()
```

```python
# make predictions for those x values and store them
preds = lm.predict(X_new)
preds
```

```python
# first, plot the observed data
bos.plot(kind='scatter', x='Crime_Rate', y='PRICE')

# then, plot the least squares line
plt.plot(X_new, preds, c='red', linewidth=2)
```

```python
import statsmodels.formula.api as smf
lm = smf.ols(formula='PRICE ~ Crime_Rate', data=bos).fit()
lm.conf_int()
```

```python
# print the p-values for the model coefficients
lm.pvalues
```

```python
# print the R-squared value for the model
lm.rsquared
```

```python
# create X and y
feature_cols = ['Crime_Rate', 'Lrg_Res_Zones', 'Industry', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
X = bos[feature_cols]
y = bos.PRICE

lm = LinearRegression()
lm.fit(X, y)

# print intercept and coefficients
print(lm.intercept_)
print(lm.coef_)
```

```python
lm = smf.ols(formula='PRICE ~ Crime_Rate + Lrg_Res_Zones + Industry + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT', data=bos).fit()
lm.conf_int()
lm.summary()
```
